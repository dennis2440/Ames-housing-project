```python
#Total rooms above ground viusalization
plt.hist(data['TotRmsAbvGrd'],bins=10,color='blue',edgecolor='black')
plt.title('Total rooms above')
plt.ylabel('frequency')
plt.xlabel('number of rooms')
plt.show()
```

```python
#visualization of saleprices
plt.hist(data['SalePrice'],bins=10,color='blue',edgecolor='black')
plt.title('Sales price Histogram')
plt.ylabel('frequency')
plt.xlabel('saleprices')

plt.tight_layout()
#plt.grid(True)
plt.show()
```

```python
#saleprice visualization in seaborn
sns.histplot(data['SalePrice'],bins=10,color='green',edgecolor='black')
plt.title('Sales price Histogram')
plt.ylabel('frequency')
plt.xlabel('number of rooms')

plt.tight_layout()
#plt.grid(True)
plt.show()
```

```python
#summary stats of saleprice
mean = np.mean(data['SalePrice'])
median = np.median(data['SalePrice'])
std_deviation = np.std(data['SalePrice'])

#print(summary stats)
print(f"Mean : {mean}")
print(f"Median : {median}")
print(f"Standard Deviation : {std_deviation}")
```

```python
#creating subset of overllcond and saleprice
subset = data[['OverallCond' ,'SalePrice' ]]
subset
```

```python
#visual distribution of the subset against saleprice
df = pd.DataFrame(data)

below_average_conditions = df[df['OverallCond']< 5]
average_conditions = df[df['OverallCond']== 5]
above_average_conditions = df[df['OverallCond']> 5]

fig, ax = plt.subplots(figsize=(20,10))

bins = range(df['SalePrice'].min(), df['SalePrice'].max(), int(df['SalePrice'].median()) //20)

ax.hist(
        x=above_average_conditions['SalePrice'],
        label= 'above average condition',
        bins= bins,
        color = 'cyan',
        alpha=0.5
)

ax.hist(
        x=average_conditions['SalePrice'],
        label= 'average condition',
        bins= bins,
        color = 'gray',
        alpha=0.7
)

ax.hist(
        x=above_average_conditions['SalePrice'],
        label= 'below_average condition',
        bins= bins,
        color = 'yellow',
        alpha=0.5
)

ax.set_title('Distributions of sales price grouped by condition ')
ax.set_ylabel('Number of Houses')
ax.set_xlabel('SalePrice')
ax.legend()
ax.grid(True , linestyle='--' , alpha=0.6)
```

```python
#Egineering and exploring a new feature
df = pd.DataFrame(data)
Age = df['YrSold'] - df['YearBuilt']
df['Age'] = Age
print(df)
```

```python
sales = data['SalePrice'].head(20)
age = df['Age'].head(20)

plt.scatter(sales,age)
plt.xlabel('Sale Price')
plt.ylabel('Age')
plt.title('Relationship Between Age and Sales price')
plt.tight_layout()
plt.show()
```

```python
#-ve correlation of age and saleprice
correlation = df['Age'].corr(df['SalePrice'])
correlation
```

```python
# using seaborn to vizualize the scatter plot of the entire dataframe
sns.scatterplot(data=df, x='SalePrice', y='Age')

plt.xlabel('Sale Price')
plt.ylabel('Age')
plt.title('Relationship Between Age and Sales Price')
plt.tight_layout()
plt.show()
```
